### Part 1. Train Test Split

```python
X = cn.drop('malignant', axis = 1)
y = cn['malignant']

y.head()
```

```python
X_train, X_test,y_train, y_test = train_test_split(
    X, 
    y, 
    test_size = 3,
    random_state = 1
)
```

we don't know if the above score is generalizable to other samples of data

In order to generate scores that we think are generalizable we will use cross-validation to generate the recall score

```python
log_reg_vanilla_cv = LogisticRegression(random_state = 1, max_iter = 10000)

log_reg_vanilla_crossval_train_mean = np.mean(
    cross_val_score(
        log_reg_vanilla_cv, #put an unfit .......
        X_train, 
        y_train, 
        scoring ='recall'
    )
)

log_reg_vanilla_crossval_train_mean
```

```python
cross_val_score(
        log_reg_vanilla_cv, #put an unfit .......
        X_train, 
        y_train, 
        scoring ='recall'
    )
```

Question: are we overfitting or underfitting?
IDK because we compare train and test metrics... but so far we only have the metric for the test folds of the data. we don't know our metrics for the train folds.

This is good , but still don't have  *training* metrics to be able to diagnose over/underfitting we'll use `cross_validate` to do so

```python
log_reg_vanilla_cv1 = LogisticRegression(random_state = 1, max_iter = 10000)

cv1 = cross_validate(
               log_reg_vanilla_cv1, #can just add a knn object here instead to do a cv for knn
               X_train,
               y_train,
               scoring = 'recall',
               return_train_score = True
            
              )
cv1
```

```python
np.mean(cv1['train_score'])
```

```python
np.mean(cv1['test_score'])
```

Through our cv process, it's not overfitting very much.
