```python
#Numpy array computations are faster than python computations as demonstrated
#below
import time

size_of_seq = 100000

def pure_python_version():
    tic = time.time()
    X = range(size_of_seq)
    Y = range(size_of_seq)
    Z = [X[i] + Y[i] for i in range(len(X)) ]
    toc = time.time()
    return toc - tic

def numpy_version():
    tic = time.time()
    X = np.arange(size_of_seq)
    Y = np.arange(size_of_seq)
    Z = X + Y 
    toc = time.time()
    return toc - tic


t1 = pure_python_version()
t2 = numpy_version()
print("python: " + str(t1), "numpy: "+ str(t2))
print("Numpy is in this example " + str(t1/t2) + " times faster!")
```

```python
## broadcasting and array math
# in Linear algebra broadcasting means can coerces two matrices of 
# diff sizes to one size such that operations can be performed

# under the hood an int or scalar is projecting to an array of len 4,
# 4 times so that the operation can happen
# Don't need to specify that an operation is an elementwise operation, numpy will do it for you.

np.array([2,3,4,6]) * 5
# this is an elementwise operation
# projected scalar 4 times to the 4 elements in the array so that the 
#operation is possible.
```

```python
li = [2,3,4,6]
li * 5
# a python list is unable to perform broadcasting, this 
# returns the list 5 times
# pandas is also able to broadcast.
```
