#### PCA Step-by-step
1. It is important to center and standardize your data. PCA lives off of correlation and covariance of your data, and using wildly different scales could lead to inflated weights for the linear combination. Let's call this centered and standardized matrix Z.
2. Calculate a covariance matrix of p x p where p responds to number of predictors.
3. Calculate the eigenvectors and eigenvalues of the covariance matrix.
4. Arrange the eigenvalues from largest to smallest. You should obtain p eigenvalues which correspond to number of components.
5. Choose the amount of components you want to include based on number of variance explained.

* Eigenvectors- are the direction of the unit scaled vector in the p-dimensional space for the principal component.(given by pca.components_)
* Eigenvalues- are the magnitude of the variation in each of the components, denoted as lambda. (given by pca.explained_variance_)
* loadings - are the covariance/ correlations between the original variables and the unit-scaled components. given by eigenvector*sqroot of eigenvalue
loadings are coefficients in linear combination describing a variable by the (standardized) components

# check out how many components there are 
There are two components that satisfy the 90% variance explainability condition

#### What are some of the advantages using PCA? 
* Some of the advantages of PCA are
-- No loss of information in dimension reduction
-- Helps reduce overfitting from massive number of features
-- since the outputted principal components are all independent, it removes any feature correlation issue
-- pca allows for easy visualization on a 2D plane using a 2D scree plot

#### What are some of the disadvantages of PCA?
-- The resulting components are not as easy interpretable as its original features
-- Data standardization is a must before PCA
-- possible loss of information from the way the number of principal components are picked



#### How does it help reduce dimensions?
* Reduces the dimensions by projecting the original features specified to lower dimensions in a way that it captures the variability in the data. PCA extracts the original features specified, and returns the linear combinations of these features called principal components. 

Principal components are the condensed forms of the original features.

## Part II. Clustering

- Describe the algorithm of [K-Means](https://www.naftaliharris.com/blog/visualizing-k-means-clustering/) Clustering 

** K-Means algorithm is an optimization algorithm which maximizes the sum total of inter-cluster distance while minimizing the sum total of intra-cluster distance. K-Means uses Euclidean distance as a distance measurement.
K-Means clustering begins with k inital clusters, and reassigns observations  to clusters, updates the center as the mean of the data points that begin to this cluster. It iterates over these steps till no improvement in centering  can be obtained.

The number of K you begin with affects clustering

Step by Step:
1. randomly instantiate k centroids
2. Assign the obs membership to centroids based on the closest distance
3. Evaluate the mean of the points on the groups and update the position of the centroid as the mean of the points
4. Reassign the membership of the datapoints based on the updated centroid
5. Keep reassign and updating position till no improvement can be found.

```python
# building a k-means clustering model 
kmeans_2 = KMeans(n_clusters=2).fit(mpg_scaled)
kmeans_3 = KMeans(n_clusters=3).fit(mpg_scaled)
kmeans_4 = KMeans(n_clusters=4).fit(mpg_scaled)
kmeans_5 = KMeans(n_clusters=5).fit(mpg_scaled)
# kmeans.fit(mpg_scaled)
# predicted = kmeans.predict(mpg_scaled) we don't do predict here since we don't have a holdout
# dataset and we have only fitted our model to the dataframe
#instead we just call the labels

kmeans_list= [kmeans_2, kmeans_3, kmeans_4, kmeans_5 ]
```

```python
# we can get the labels
# kmeans.labels_
```

```python
# what are some of the ways to evaluate k-means?
from sklearn.metrics import calinski_harabasz_score
```

```python
CH_score= []
labels_list = []

for model in kmeans_list:
    labels = model.labels_
    CH_score.append(calinski_harabasz_score(mpg_scaled, labels) )
    labels_list.append(labels)
CH_score
```

```python
plt.plot([2, 3, 4, 5], CH_score)
plt.xticks([2, 3,4,5])
plt.title('Calinski Harabasz Scores for Different Values of K')
plt.ylabel('Variance Ratio')
plt.xlabel('K=')
plt.show()
```

```python
# define our x and y
X = df.review
y = df.label
```

```python
# what are some of the steps to engineer the features?

# stop words removal

stopwords = ['am','he','i','the','hi']

# what if you need to remove more stopwords than this collection?

# use tfidf, with unigrams and bigrams, and remove stop words
vectorizer = TfidfVectorizer(ngram_range=(1,2), stop_words=stopwords)

## could you add some optional parameters to that?
```

**Interpretation of TFIDF** 

What does a high value of TFIDF tell you about a certain word?
A high TFIDF indicates that the word is important

```python
# what are the step by step process of doing text classification

# step 1 - train test split
X_train, X_test, y_train, y_test = train_test_split(X, y )

# step 2 - tokenize and feature engineer using tfidf
# here without specifying a separate tokenizer, tokenization happens within the vectorizer, 
# creating the training and testing features, which will then be used in training the model as features

train_features = vectorizer.fit_transform(X_train)
test_features = vectorizer.transform(X_test)

# step 3 - fit the model to the training set, test the model on the testing set
classifier = RandomForestClassifier() # pass your own params
classifier.fit(train_features, y_train)

# step 4 - output prediction on testing set, compare true labels to predicted labels, get accuracy
pred = classifier.predict(test_features)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(pred, y_test)
```

# What is stationarity and does this satisfy the stationarity assumption

```python
# do a dickey fuller test to find out whether our dataset is stationary or not
from statsmodels.tsa.stattools import adfuller
test = adfuller(temp_monthly['Temp'])
dfoutput = pd.Series(test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
print(dfoutput)
```

based on the df test.... not stationary.... p-value is greater.
The more obs you have the more like a significance out of chance... 
decreasing the freq to monthly, we reduced any noise. Here, we reject the null of stationarity and state that they is no significant evidence of stationarity at the 95% confidence because the p value is more than 0.05.
